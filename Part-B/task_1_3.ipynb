### 1. Baseline method

#### **Baseline**
The paper mainly compares AdaGrad with a basic learning method called **standard stochastic gradient descent**, especially Online Gradient Descent (Zinkevich's algorithm). This is a simple and commonly used method where the model updates its parameters step by step while learning from the data.

### 2. Limitation of baseline

#### **Limitation**
Standard gradient descent uses one **fixed learning rate** for all features in the dataset. This means:
- A feature that appears millions of times
- And a feature that appears only once

Both get exactly the same learning rate. Because of this, the model often struggles to properly learn from rare but important features, since they are treated the same as very common features.

### 3. How paper fixes it

#### **Improvement**
AdaGrad solves this problem by giving **each feature its own learning rate**. It calculates the learning rate based on how often that feature has appeared before. So what happens is:
- **Features that appear very frequently** $\rightarrow$ learning rate becomes smaller
- **Features that appear rarely** $\rightarrow$ learning rate becomes larger

This forces the model to pay more attention to rare but important features, helping it learn better.

### 4. When paper might fail

#### **Failure Scenario**
AdaGrad may not perform better than the baseline if the dataset is **dense and uniform**. This means:
- Almost every feature appears in every data row
- All features have similar importance and scale
- There are no rare features

Since AdaGrad’s main advantage is adjusting learning rates between rare and common features, this advantage disappears in such datasets. In this situation:
- All feature learning rates shrink at almost the same speed
- AdaGrad behaves very similar to standard gradient descent

However, AdaGrad will still be slower and use more memory, because it has to continuously store and calculate the past squared gradients for every feature.
