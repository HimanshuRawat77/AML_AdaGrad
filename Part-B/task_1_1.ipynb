Step 2: Accumulate Historical Gradients

Description:
Instead of just looking at the current error, the algorithm takes the gradient for each individual feature, squares it, and adds it to a running historical total of all past squared gradients for that exact feature.

Purpose:
To keep a permanent memory of how frequently and heavily each feature has been updated over time; frequently seen features build up a massive total, while rare features keep a small total.

Step 3: Calculate the Adaptive Scaling Matrix

Description:
The algorithm creates a custom scaling factor for every single feature by taking the square root of its historical sum (from Step 2) and adding a tiny smoothing number ($\delta$) to guarantee the math never accidentally divides by zero.

Purpose:
To prepare a custom, per-feature adjustment for the learning rate that will force the step size to shrink down for common features while staying large for rare features.

Step 4: Update the Model Weights

Description:
The model finally updates its weights by stepping in the direction that reduces the error, but it divides the global step size (learning rate) by the custom scaling factor calculated in Step 3 for each specific feature.
