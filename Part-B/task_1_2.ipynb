### Assumption 1

#### **Assumption**
The functions we are trying to optimize (loss functions) are **convex**, meaning they look like a bowl shape.

#### **Why the method needs it**
The mathematical proof that AdaGrad works depends on the loss function being shaped like a smooth bowl.
If the surface is bowl-shaped:
- There is only one best solution at the bottom.
- The algorithm can easily slide down to that lowest point.
- It won’t get stuck anywhere else.

So the theory of the paper assumes the path to the best solution is smooth and predictable.

#### **Violation scenario**
In real life, especially when training deep neural networks, the loss function is not bowl-shaped. Instead it looks like a mountain range with many valleys and bumps. Because of this:
- The algorithm might get stuck in a local minimum (a fake bottom).
- The guarantees given in the paper may not work anymore.

#### **Paper reference**
Section 1.1 explains this mathematically:

Each function is of the form $\phi_t(x) = f_t(x) + \phi(x)$ where both functions are convex.

### Assumption 2

#### **Assumption**
Rare features in the dataset (features that appear very rarely) are very useful and meaningful for making predictions.

#### **Why the method needs it**
AdaGrad works by adjusting learning rates differently for different features.
- Features that appear **often** $\rightarrow$ learning rate becomes smaller.
- Features that appear **rarely** $\rightarrow$ learning rate becomes larger.

The idea is that rare features might carry important information, so AdaGrad tries to learn from them faster. But this only works if rare features are actually useful.

#### **Violation scenario**
Sometimes rare features are not useful at all. For example:
- Random typing mistakes in text data
- Broken sensor readings
- Data entry errors
- Random noise in the dataset

If AdaGrad increases the learning rate for these noisy features, it will amplify the noise, which can damage the model’s accuracy.

#### **Paper reference**
Section 1 (Introduction) mentions this idea:
"Infrequently occurring features are often highly informative and discriminative."

### Assumption 3

#### **Assumption**
The gradients (error signals used to update the model) must stay within a reasonable limit and should not become extremely large.

#### **Why the method needs it**
AdaGrad updates the learning rate using this idea:
1. It keeps adding the squares of all past gradients.
2. This sum is placed in the denominator of the learning rate formula.

If a gradient suddenly becomes very large, then:
- The denominator becomes huge.
- The learning rate becomes extremely small.

When this happens, the model almost stops learning for that feature, because the updates become nearly zero.

#### **Violation scenario**
This problem often happens when training Recurrent Neural Networks (RNNs) on time-series data. They sometimes suffer from the **exploding gradient problem**, where:
- Error values grow extremely large.
- Gradients blow up.

When this happens, AdaGrad’s learning rate can shrink so much that the model cannot learn anymore.

#### **Paper reference**
Section 3 (Theorem 5) mentions this mathematically:
The proofs assume that gradients have a maximum limit:
$\delta \ge \max_t ||g_t||_\infty$

This means the gradient values must not exceed a certain bound.
